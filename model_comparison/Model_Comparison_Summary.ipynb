```python
# Import pandas for tables
import pandas as pd

# Import matplotlib for charts
import matplotlib.pyplot as plt
```

```python
# Real model comparison results from your notebooks

results = [
    {
        "Model": "Decision Tree",
        "Accuracy": 0.4167,
        "Precision": 0.4447,
        "Recall": 0.4167,
        "F1-score": 0.3859
    },
    {
        "Model": "Logistic Regression",
        "Accuracy": 0.7083,
        "Precision": 0.6944,
        "Recall": 0.7083,
        "F1-score": 0.6869
    },
    {
        "Model": "Random Forest",
        "Accuracy": 0.6250,
        "Precision": 0.6269,
        "Recall": 0.6250,
        "F1-score": 0.6071
    },
    {
        "Model": "SVM",
        "Accuracy": 0.6250,
        "Precision": 0.6190,
        "Recall": 0.6250,
        "F1-score": 0.6051
    }
]

comparison_df = pd.DataFrame(results)

comparison_df
```

```python
# Sort models from highest accuracy to lowest accuracy

comparison_df_sorted = comparison_df.sort_values(
    by="Accuracy",
    ascending=False
)

comparison_df_sorted
```

```python
# Draw bar chart comparing model accuracy

plt.figure(figsize=(9, 5))

plt.bar(
    comparison_df["Model"],
    comparison_df["Accuracy"]
)

plt.title("Model Accuracy Comparison")
plt.xlabel("Machine Learning Models")
plt.ylabel("Accuracy")
plt.ylim(0, 1)

plt.xticks(rotation=20)
plt.show()
```

```python
# Draw bar chart comparing F1-score

plt.figure(figsize=(9, 5))

plt.bar(
    comparison_df["Model"],
    comparison_df["F1-score"]
)

plt.title("Model F1-score Comparison")
plt.xlabel("Machine Learning Models")
plt.ylabel("F1-score")
plt.ylim(0, 1)

plt.xticks(rotation=20)
plt.show()
```

```python
# Compare Precision and Recall together

comparison_df.set_index("Model")[["Precision", "Recall"]].plot(
    kind="bar",
    figsize=(9, 5)
)

plt.title("Precision and Recall Comparison")
plt.xlabel("Machine Learning Models")
plt.ylabel("Score")
plt.ylim(0, 1)

plt.xticks(rotation=20)
plt.show()
```

```python
# Select the model with the highest accuracy

best_model = comparison_df.loc[
    comparison_df["Accuracy"].idxmax()
]

print("Best Performing Model Based on Accuracy:")
print(best_model)
```
